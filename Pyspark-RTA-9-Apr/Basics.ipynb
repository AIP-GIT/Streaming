```python
from pyspark import SparkConf, SparkContext
conf = SparkConf().setMaster("local").setAppName("WordCount")
sc = SparkContext(conf = conf)
```

```python
rdd = sc.parallelize([1,2,3,4])
```

```python
def squareIt(x):
    return x*x 
```

```python
rdd = rdd.map(lambda x:x*x)
```

```python
rdd.map(squareIt)
```

```python
rdd.collect()
```
