# Weather

```python
lines = sc.textFile(r"/home/ubuntu-poc/workspace/projects/Streaming/PySpark-Basics/data/weather.csv")
```

```python
def parseLine(line):    
    fields = line.split(',')    
    stationID = fields[0]    
    entryType = fields[2]
    # Convert the temperature to farenheit
    temperature = float(fields[3]) * 0.1 * (9.0 / 5.0) + 32.0    
    return (stationID, entryType, temperature)
```

```python
parsedLines = lines.map(parseLine)
```

```python
parsedLines.count()
```

```python
sample = parsedLines.sample(withReplacement=False,fraction=.20)
```

```python
sample.count()
```

```python
372/1825 
```

```python
sample.collect()
```

```python
# Filter all the line with TMIN

minTemps = parsedLines.filter(lambda x: "TMIN" in x[1])
```

```python
minTemps.count()
```

```python
stationTemps = minTemps.map(lambda x:(x[0],x[2]))
stationTemps.top(5)
```

```python
stationTemps.count()
```

```python
# Find Min temp by station ID
overallMinStation = stationTemps.reduceByKey(lambda x,y:min(x,y))
```

```python
overallMinStation.count()
```

```python
stationTemps.countByKey()
```

```python
for result in overallMinStation.collect():
    print(result)
```

```python
# code for Max temperature
maxTemps = parsedLines.filter(lambda x: "TMAX" in x[1])
stationTemps = maxTemps.map(lambda x:(x[0],x[2]))
overallMaxStation = stationTemps.reduceByKey(lambda x,y:max(x,y))

for result in overallMaxStation.collect():
    print(result)
```

# Count Word Occurance

```python
lines = sc.textFile(r"/home/ubuntu-poc/workspace/projects/Streaming/PySpark-Basics/data/book.txt")
```

```python
words = lines.flatMap(lambda x:x.split())
words.top(10)
```

```python
wordCounts = words.countByValue()
```

```python
for word,count in wordCounts.items():
    cleanWord = word.encode('ascii','ignore')
    if(cleanWord):
        print(cleanWord.decode()+" "+str(count))
```

## Normalize words

```python
import re 

def normalizeWords(text):
    return re.compile(r'\W+',re.UNICODE).split(text.lower())
```

```python
words = lines.flatMap(normalizeWords)
wordCounts = words.countByValue()
```

```python
for word,count in wordCounts.items():
    cleanWord = word.encode('ascii','ignore')
    if(cleanWord):
        print(cleanWord.decode()+" "+str(count))
```

```python
# sort

##below is equivalent to countBYyValue() function
wordCounts = words.map(lambda x:(x,1)).reduceByKey(lambda x,y:x+y)
##sorting with a hack
wordCountsSorted = wordCounts.map(lambda x:(x[1],x[0])).sortByKey()
```

```python
type(wordCountsSorted)
```

```python
for result in wordCountsSorted.collect():
    count = str(result[0])
    word = result[1].encode('ascii','ignore')
    if(word):
        print(word.decode()+":\t"+count)
```

# Total Amount spent by the customer

```python
# customer ID, an item ID, and the amount spent on that item. 
lines = sc.textFile(r"/home/ubuntu-poc/workspace/projects/Streaming/PySpark-Basics/data/customer-orders.csv")
```

```python
def extractCustomerPricePairs(line):
    fields = line.split(',')
    return (int(fields[0]),float(fields[2]))
```

```python
mappedRdd = lines.map(extractCustomerPricePairs)
```

```python
mappedRdd.top(5)
```

```python
totalByCustomer = mappedRdd.reduceByKey(lambda x,y:x+y)
```

```python
for result in totalByCustomer.collect():
    print(result)
```

```python
# flip and sort

flipped = totalByCustomer.map(lambda x:(x[1],x[0]))
totalByCustomerSorted = flipped.sortByKey()
```

```python
for result in totalByCustomerSorted.collect():
    print(result)
```
